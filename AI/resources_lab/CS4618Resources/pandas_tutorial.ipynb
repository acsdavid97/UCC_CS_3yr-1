```python
df[df['type'] == 'Apartment']
```

```python
# & | ~ -> and or not
# extract apartments or teraced houses
df[(df['type'] == 'Apartment') | (df['type'] == 'Terraced')]
```

```python
# how many apartments
df[df['type'] == 'Apartment'].shape
len(df[df['type'] == 'Apartment'])
# fastest because we don't need to extract the actual columns
(df['type'] == 'Apartment').sum()
```

```python
m, n = df.shape
```

```python
# permuation from 0 to m
np.random.permutation(m)
```

```python
# shuffle (reorder) randomly the dataset
df = df.take(np.random.permutation(m))
```
